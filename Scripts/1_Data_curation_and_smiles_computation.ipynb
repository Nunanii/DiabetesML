Compute SMILES from PubChem CID

```python
data_4 = data_3.dropna(subset=['cid'])
data_4['cid'].dtypes

data_4['cid'] = data_4['cid'].apply(np.int64)

data_4['cid'].dtypes #convert to integer
```

```python
pubchem_cid = data_4['cid']
print(str(len(pubchem_cid)))
```

```python
data_4
```

```python
Smiles_1 = []      #create an empty list to store SMILES
for a in pubchem_cid:
    result_1 = pcp.Compound.from_cid(a).isomeric_smiles  #compute SMILES
    Smiles_1.append(result_1)
    tm.sleep(0.1)  #provide delay
```

```python
Smiles_1 #list of smiles computed
```

Add SMILES to original file

```python
data_4.insert(1, 'Smiles', Smiles_1)

data_4['activity'] = 'Active'
data_4 = data_4.drop('cmpdname', axis=1)
validation = data_4.rename({'cid':'PUBCHEM_CID', 'Smiles':'PUBCHEM_EXT_DATASOURCE_SMILES', 'activity':'PUBCHEM_ACTIVITY_OUTCOME'}, axis=1)
```

```python
validation.to_csv('/Manuscrip_preparation/compounds/validation.csv')
```

```python
inactive_subset = inactive.sample(n=700, random_state = 42)  #randomly select 700 inactives for model training
inactive_subset.to_csv('/Manuscrip_preparation/compounds/inactive_subset2.csv')
```

```python
#merge 1000 inactive compounds with available actives
main_data = pd.concat([active, inactive_subset], ignore_index=True, sort=False)
main_data.to_csv('Manuscrip_preparation/compounds/GLP1R_data2.csv')
```
